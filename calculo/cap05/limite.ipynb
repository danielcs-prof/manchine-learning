```python
import numpy as np
import matplotlib.pyplot as plt

# Função definida com a remoção da indeterminação
def f(x):
    return (x**2 - 1)/(x - 1)

# Valores de x se aproximando de 1 pela esquerda e pela direita
x_left = np.linspace(0.9, 0.9999, 10)
x_right = np.linspace(1.0001, 1.1, 10)

# Calculando os valores de f(x)
y_left = f(x_left)
y_right = f(x_right)

# Imprimindo as aproximações
print("Aproximação pela esquerda:")
for x_val, y_val in zip(x_left, y_left):
    print(f"x = {x_val:.5f}, f(x) = {y_val:.5f}")

print("\nAproximação pela direita:")
for x_val, y_val in zip(x_right, y_right):
    print(f"x = {x_val:.5f}, f(x) = {y_val:.5f}")

# Plotando o gráfico
x_all = np.concatenate((x_left, x_right))
y_all = f(x_all)

plt.plot(x_all, y_all, 'bo', label='Aproximações')
plt.axhline(y=2, color='r', linestyle='--', label='Limite esperado (y=2)')
plt.axvline(x=1, color='gray', linestyle=':', label='x = 1')
plt.title('Aproximação Numérica do Limite')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
