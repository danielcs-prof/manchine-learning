```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def planoCartesiano(xMax,yMax,title='Plano Cartesiano'):
    
    fig, ax = plt.subplots(figsize=(5,5))       # Cria o gráfico e eixos
    ax.spines['top'].set_color('none')          # Oculta o eixo localizado no Topo
    ax.spines['right'].set_color('none')        # Oculta o eixo localizado à direita
    ax.spines['left'].set_position('zero')      # Desloca o eixo da esquerda para o ponto zero
    ax.spines['bottom'].set_position('zero')    # Desloca o eixo inferior para o ponto zero
    ax.set_xlim(-xMax,xMax)                     # Defne limites do eixo x: (inferior, superior)
    ax.set_ylim(-yMax,yMax)                     # Defne limites do eixo y: (inferior, superior)
    ax.set_xticks(np.arange(-xMax,xMax,1))  #define os pontos de marcação do eixo x
    ax.set_yticks(np.arange(-yMax,yMax,1))  #define os pontos de marcação do eixo y
    ax.grid(True,linestyle='--')                # Habilita o grid do gráfico
    ax.set_xlabel('x',x=0.95)                   # Atribui um nome ao eixo x
    ax.set_ylabel('y',y=0.95)                   # Atribui um nome ao eixo y
    ax.set_title(title)            # Atribui um título ao gráfico
    ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False) # atribui seta ao eixo X
    ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False) # atribui seta ao eixo Y
    return fig, ax

# Gerando os pontos das coordenadas (x,y)
pontos = [(2,3),(-3,2),(-2,-3),(3,-4)]          # Cria um vetor com quatro tuplas - coordenadas (x,y)
x = [cx[0] for cx in pontos]                    # Cria um vetor que filtra somente o x do vetor de tuplas
y = [cy[1] for cy in pontos]                    # Cria um vetor que filtra somente o y do vetor de tuplas

fig, ax = planoCartesiano(5,5)                  # Chama a função planoCartesiano para gerar o plano cartesiano
plt.plot(x,y,'o')                               # Plota cada ponto do ventor pontos desmembrado em x e y
plt.show()
```

```python
# Distância entre dois pontos P1 e P2
def distancia(p1,p2):
    return np.sqrt( (p2[0]-p1[0])**2 + (p2[1]-p1[1])**2 )

# Ponto médio entre dois pontos P1 e P2
def pontoMedio(p1,p2):
    pmx = (p2[0] + p1[0])/2
    pmy = (p2[1] + p1[1])/2
    return (pmx,pmy)

# Gerando os pontos P1 e P2
pontos = [(1,4),(4,1)]

# Calcula a ditância entre P1 e P2
d = distancia(pontos[0],pontos[1])

# Calcular o ponto médio entre P1 e P2
pm = pontoMedio(pontos[0],pontos[1])

x = [cx[0] for cx in pontos]     # Cria um vetor que filtra somente o x do vetor de tuplas
y = [cy[1] for cy in pontos]     # Cria um vetor que filtra somente o y do vetor de tuplas

fig, ax = planoCartesiano(5,5,'Distância e Ponto Médio')   # Chama a função planoCartesiano para gerar o plano cartesiano

# Plota os pontos P1, P2 e o ponto médio
plt.plot(x,y,marker='o')         
plt.plot(pm[0],pm[1],'o',color='red')

# Imprime o texto que identifica cada ponto
ax.text(x[0],y[0]+0.4,f'P1=({x[0]},{y[0]})')
ax.text(x[1],y[1]-0.5,f'P2=({x[1]},{y[1]})')
ax.text(pm[0]-2,pm[1]-0.9,f'Ponto médio\n({pm[0]},{pm[1]})')

plt.show()
```

```python
# Translação de pontos
def translacao(pontos,transformacao):
    x = np.array([cx[0] + transformacao[0] for cx in pontos])
    y = np.array([cy[1] + transformacao[1] for cy in pontos])
    return list( zip(x,y) )

# Gera 4 pontos do quadrado em um vetor de tuplas
pontos = [(1,1),(1,4),(3,4),(3,1),(1,1)]
x = [cx[0] for cx in pontos]    # Separa as coordenadas x das tuplas
y = [cy[1] for cy in pontos]    # Separa as coordenadas y das tuplas

# Vetor de tuplas com os pontos transladados
pontosT = translacao(pontos,(-5,-5))   # Calcula a translação
xt = [cx[0] for cx in pontosT]    # Separa as coordenadas x das tuplas
yt = [cy[1] for cy in pontosT]    # Separa as coordenadas y das tuplas

# Exibe o gráfico
fig, ax = planoCartesiano(5,5,'Translação de Pontos')
plt.plot(x,y,marker='o') # Pontos sem translação
plt.plot(xt,yt,marker='o',color='red') # Pontos transladados
plt.show()
```
