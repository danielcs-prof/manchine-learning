```python
import numpy as np
import matplotlib.pyplot as plt

# Função de custo e derivada
f = lambda w: (w - 3)**2
df = lambda w: 2 * (w - 3)

# Inicialização
w = 0
lr = 0.1
hist_w = [w]
hist_f = [f(w)]

# Executa 20 iterações
for _ in range(20):
    grad = df(w)
    w = w - lr * grad
    hist_w.append(w)
    hist_f.append(f(w))

# Gráfico da função e dos passos do gradiente
w_vals = np.linspace(-1, 6, 200)
f_vals = f(w_vals)

plt.figure(figsize=(8,5))
plt.plot(w_vals, f_vals, label='Função de Custo')
plt.scatter(hist_w, hist_f, c='red', label='Passos do Gradiente')
plt.plot(hist_w, hist_f, 'r--')
plt.title('Gradiente Descendente na função f(w) = (w - 3)^2')
plt.xlabel('w')
plt.ylabel('f(w)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
