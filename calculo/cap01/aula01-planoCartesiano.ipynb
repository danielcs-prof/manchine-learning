## 1.1 - Plano Cartesiano

<p>
    O plano cartesiano é um sistema de coordenadas bidimensional composto por duas retas perpendiculares entre si, chamadas de eixo x (horizontal) e eixo y (vertical), que se cruzam na origem (ponto de coordenadas (0,0)), dividindo o plano em quatro regiões chamadas quadrantes; nesse sistema, cada ponto é identificado por um par ordenado (x, y), onde x representa a distância horizontal em relação ao eixo y e y representa a distância vertical em relação ao eixo x, permitindo a representação gráfica de relações entre duas variáveis e a visualização de conceitos matemáticos fundamentais como distância, ponto médio e alinhamento.
</p>
### EXEMPLO 01: Plano Cartesiano, pontos e quadrantes
<p>
    Considere a tabela abaixo com as coordenadas (x,y) para 4 pontos:
</p>

| Variável | Ponto 1 | Ponto 2 | Ponto 3 | Ponto 4 |
|:---------|:--------:|:--------:|:--------:|---------:|
| **x**        |     3    |    -3    |    -3    |        3 |
| **y**        |     3    |    -3    |    -3    |        3 |
<p>
    O código Python exibe o gráfico de dispersão para os pontos
</p>

```python
# PALNO CARTEZIANO
import numpy as np
import matplotlib.pyplot as plt

#1 - GERAR OU CARREGAR OS DADOS
coordenadas = [(3,3), (-3,3), (-3,-3), (3,-3)]
x = [ponto[0] for ponto in coordenadas]
y = [ponto[1] for ponto in coordenadas]

#2 - CÁLCULO

#3 - CRIA O GRÁFICOS E EIXOS
fig, ax = plt.subplots( figsize=(5,5) )

#4 - TITULO E LEGENDA
plt.title('Plano Cartesiano')

#5 - CONFIGURA OS EIXOS
ax.spines['top'].set_color('none')         # oculta o eixo top
ax.spines['right'].set_color('none')       # oculta o eixo right
ax.spines['left'].set_position('zero')     # desloca o eixo para o ponto 0
ax.spines['bottom'].set_position('zero')   # desloca o eixo para o ponto 0
ax.set_xlabel('Eixo x', x=0.95)            # atribui e posiciona o nome ao eixo
ax.set_ylabel('Eixo y', y=0.95)            # atribui e posiciona o nome ao eixo 
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False) # atribui seta ao eixo X
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False) # atribui seta ao eixo Y

#6 - CONFIGURAÇÃO DO GRID
valor_max = max(np.max(x),np.max(y)) + 3    # localiza o maio valo de X e Y e depois de ambos
ax.set_xlim(-valor_max,valor_max)           # define o limite inferior e superior do eixo X
ax.set_ylim(-valor_max,valor_max)           # define o limite inferior e superior do eixo Y
ax.set_xticks(np.arange(-valor_max,valor_max,1))
ax.set_yticks(np.arange(-valor_max,valor_max,1))
ax.grid(True, linestyle='--')              # habilita o grid

#6 - PLOTAR OS DADOS
ax.scatter(x,y)                             # plota gráfico e dispersão (pontos isolados)
for i, ponto in enumerate(coordenadas):
    plt.text(ponto[0]+0.2,ponto[1]+0.2, f'({ponto[0]},{ponto[1]})')
    plt.text(ponto[0]-1,ponto[1]-1, f'Quadrante {i+1}')
    
#7 - EXIBIR OS GRÁFICOS
plt.show()

#8 - EXIBIR RESULTADO NO CONSOLE
print()
print(f'Coordenada X: {x}')
print(f'Coordenada Y: {y}')
print()
```

##1.2 - Gráfico de Dispersão

<p>
    Um gráfico de dispersão é uma representação gráfica que exibe pares de valores de duas variáveis em um plano cartesiano, onde cada ponto no gráfico corresponde a um par ordenado (x, y), permitindo visualizar a distribuição dos dados e identificar possíveis relações, padrões ou tendências entre as variáveis analisadas, sendo especialmente útil para mostrar correlações ou a dispersão dos dados em torno de uma relação esperada.
</p>

###EXEMPLO 02: Gráfico de dispersão
<p>
    O número E (em milhões de pessoas) dos trabalhadores do setor privado nos Estados Unidos entre 2000 e 2009 é mostrado na tabela, em que t representa o ano. Esboce um gráfico de dispersão dos dados. <i>(Fonte: US Bureau of Labor Statistics)</i>
</p>

| Variável | 2000 | 2001 | 2002 | 2003 | 2004 | 2005 | 2006 | 2007 | 2008 | 2009 |
|:--------:|:----:|:----:|:----:|:----:|:----:|:----:|:----:|:----:|:----:|:----:|
| **E**    |  111 |  111 |  109 |  108 |  110 |  112 |  114 |  115 |  114 |  108 |

<p>
    O código Python para exibir o gráfico de dispersão está abaixo.
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

#1 - GERAR OU CARREGAR OS DADOS

anos = np.arange(2000,2010,1)
qtTrabalhadoresPrivados = [111,111,109,108,110,112,114,115,114,108]

#2 - CALCULAR

#3 - CRIAR O GRÁFICO E EIXOS
fig, ax = plt.subplots( figsize=(10,5) )

#4 - TITULOS E LEGENDAS
plt.title('Total de Empregos Privados nos EUA')

#5 - CONFIGURAR OS EIXOS
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xlabel('t (Anos)')
ax.set_ylabel('Empregados Privados (E)- Milhões)')

#5 - CONFIGURAR GRID E ESCALA
ax.set_xlim(np.min(anos)-1, np.max(anos)+1)
ax.set_ylim(np.min(qtTrabalhadoresPrivados)-6,np.max(qtTrabalhadoresPrivados)+1)
ax.set_xticks(np.arange(np.min(anos)-1, np.max(anos)+2,1))
ax.set_yticks(np.arange(np.min(qtTrabalhadoresPrivados)-6, np.max(qtTrabalhadoresPrivados)+2,1))
ax.grid(True, linestyle='--')


#6 - PLOTAR OS DADOS
ax.scatter(anos,qtTrabalhadoresPrivados)


#7 - EXIBIR OS DADOS
plt.show()
print(anos)
print(qtTrabalhadoresPrivados)
```

##1.3 - Distância entre dois pontos
<p>
A <b>distância entre dois pontos</b> no plano cartesiano é uma medida que expressa o comprimento do segmento de reta que os conecta, calculada pela fórmula:    
</p>

$$
d = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2},
$$
<p>
onde $(x_1, y_1)$ e $(x_2, y_2)$ são as coordenadas dos pontos, sendo essa fórmula derivada do teorema de Pitágoras e permitindo determinar a separação exata entre quaisquer dois pontos no plano.    
</p>


```python
import numpy as np
import matplotlib.pyplot as plt

#1 - GERANDO OU CARREGANDO DADOS
coordenadas = [(2,3),(7,8)]
x = [pontos[0] for pontos in coordenadas]
y = [pontos[1] for pontos in coordenadas]

#2 - CÁLCULO
d = np.sqrt( np.pow(x[0]-x[1],2) + np.pow(y[0]-y[1],2) )

#3 - CRIAR O GRÁFICO E EIXOS
fig, ax = plt.subplots( figsize=(5,5) )

#4 - CRIAR TITULOS E LEGENDAS
plt.title('Distância entre dois pontos')

#5 - CONFIGURAR OS EIXOS
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.set_xlabel('x', x=0.95)
ax.set_ylabel('y', y=0.95)

#5 - CONFIGURACAO DE GRID E ESCALA
ax.set_xlim( 0, np.max(x)+2)
ax.set_ylim( 0,  np.max(y)+2)
ax.set_xticks(np.arange(0, np.max(x)+2,1))
ax.set_yticks(np.arange(0, np.max(y)+2,1))
ax.grid(True, linestyle='--')

#6 - PLOTAR OS DADOS
plt.plot(x,y, marker='o')
ax.text(x[0],y[0]-0.5,f'({x[0]},{y[0]})')
ax.text(x[1],y[1]-0.5,f'({x[1]},{y[1]})')

#6 - EXIBIR GRÁFICOS
plt.show()

print(x)
print(y)
```
